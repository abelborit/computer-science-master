```python
import matplotlib.pyplot as plt
import numpy as np

entradaDatos = ["10","20", "40", "80", "160","320", "640", "1280", "2560","5120"]
tiempoProcesamientoAlgoritmoGenetico = [0.001149, 0.001898, 0.003752, 0.010151, 0.032446, 0.117269, 0.386897, 1.447995, 5.582167, 21.98101]

# Aumentar el tamao del grfico
plt.figure(figsize=(12,8))

# Introducir series de datos con color, marcador, etiqueta y estilo
plt.plot(entradaDatos, tiempoProcesamientoAlgoritmoGenetico, label="Algoritmo Genetico")

# Activar leyendas
plt.legend()

# Titulo de la grfica
plt.title("Tiempo procesamiento para el problema del Agente Viajero aplicando Algoritmo Genetico")

# Etiquetado de los eje
plt.xlabel("Cantidad de nodos")
plt.ylabel("Tiempo de procesamiento (milisegundos)")

# Personalizar Eje de Puntaje de 0 a 28 en incrementos de 2
plt.yticks(np.arange(0, 28, 2))

# Activar cuadrcula
plt.grid(alpha = 0.3)

plt.show()
```
