```python
import matplotlib.pyplot as plt
import numpy as np

entradaDatos = ["10","20", "40", "80", "160","320", "640", "1280", "2560","5120"]
tiempoProcesamientoMST = [0.0000, 0.0000, 0.0000, 0.0000, 0.0000, 0.5024, 0.6837, 4.0896, 12.5886, 33.6501]

# Aumentar el tamao del grfico
plt.figure(figsize=(12,8))

# Introducir series de datos con color, marcador, etiqueta y estilo
plt.plot(entradaDatos, tiempoProcesamientoMST, label="MST")

# Activar leyendas
plt.legend()

# Titulo de la grfica
plt.title("Tiempo procesamiento para el problema del Agente Viajero aplicando Minimum Spanning Tree")

# Etiquetado de los eje
plt.xlabel("Cantidad de nodos")
plt.ylabel("Tiempo de procesamiento (milisegundos)")

# Personalizar Eje de Puntaje de 0 a 42 en incrementos de 2
plt.yticks(np.arange(0, 42, 2))

# Activar cuadrcula
plt.grid(alpha = 0.3)

plt.show()
```
