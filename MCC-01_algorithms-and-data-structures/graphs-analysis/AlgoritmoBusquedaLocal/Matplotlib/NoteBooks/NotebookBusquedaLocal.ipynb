```python
import matplotlib.pyplot as plt
import numpy as np

entradaDatos = ["10","20", "40", "80", "160","320", "640", "1280", "2560","5120"]
tiempoProcesamientoBusquedaLocal = [4.2656, 7.1241, 13.8356, 36.5711, 84.2718, 293.8134, 853.6143, 2722.0933, 8907.4015, 25858.1051]

# Aumentar el tamao del grfico
plt.figure(figsize=(12,8))

# Introducir series de datos con color, marcador, etiqueta y estilo
plt.plot(entradaDatos, tiempoProcesamientoBusquedaLocal, label="Busqueda Local")

# Activar leyendas
plt.legend()

# Titulo de la grfica
plt.title("Tiempo procesamiento para el problema del Agente Viajero aplicando Busqueda Local")

# Etiquetado de los eje
plt.xlabel("Cantidad de nodos")
plt.ylabel("Tiempo de procesamiento (milisegundos)")

# Personalizar Eje de Puntaje de 0 a 31000 en incrementos de 1500
plt.yticks(np.arange(0, 31000, 1500))

# Activar cuadrcula
plt.grid(alpha = 0.3)

plt.show()
```
